```python
#2. dictionary:
#a. Buat dictionary keranjang dengan data: {"apel": 5000, "pisang": 3000, "jeruk": 7000, "anggur": 15000}
keranjang = {
    "apel" : 5000,
    "pisang" : 3000,
    "jeruk" : 7000,
    "anggur" : 15000
}
#b. Gunakan perulangan untuk menghitung total harga semua barang
total = 0
for buah, harga in keranjang.items():
  total += harga

#c. Tampilkan setiap barang dan harganya dalam format: ex: "Harga apel: Rp 5000"
  print(f"Harga {buah}\t = Rp{harga}")

#d. Tampilkan harga total belanja
print("Total harga belanja = Rp", total)
```
